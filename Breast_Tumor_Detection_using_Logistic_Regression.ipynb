```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection  import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
df=sklearn.datasets.load_breast_cancer()
df
```

```python
data_frame=pd.DataFrame(df.data,columns=df.feature_names)
data_frame
```

```python
data_frame['label']=df.target
```

```python
data_frame.head()
```

```python
data_frame.info()
```

```python
data_frame.shape
```

```python
data_frame.describe()
```

```python
data_frame['label'].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
X=data_frame.drop(columns='label',axis=1)
Y=data_frame['label']
```

```python
print(X)
```

```python
[X_train,X_test,Y_train,Y_test]=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
```

```python
print("Accuracy on training data =",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
print("Accuracy on training data =",test_data_accuracy)
```

```python
cm=confusion_matrix(Y_test,X_test_prediction)
sns.heatmap(cm,annot=True,fmt='g',xticklabels=['malignant','benign'],yticklabels=['malignant','benign'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=13)
plt.show()
```

```python
accuracy=accuracy_score(Y_test,X_test_prediction)
print("Accuracy",accuracy)
precision=precision_score(Y_test,X_test_prediction)
print("Precision",precision)
recall=recall_score(Y_test,X_test_prediction)
print("Recall Score",recall)
F1_score=f1_score(Y_test,X_test_prediction)
print("F1_score",F1_score)
```
